Q2. What is the difference between remove() and pop() functions in Python?

A2. The difference between remove() and pop() in Python is that remove() removes the first occurrence of a specified element from the list, while pop() removes an element from a specific index and returns it.

Q4. Write a Python code to count the number of occurrences of an element in a list.

A4. 

elements = ["a", "b", "c", "a", "d", "a"]

element = "a"

count = elements.count(element)

print(element, "occurs", count, "times in the list")
