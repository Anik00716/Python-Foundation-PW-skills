Q1. What is a tuple in Python?

A1. In Python, a tuple is an ordered collection of elements, similar to a list. However, tuples are immutable, which means they cannot be modified once they are created.

Q2. How do you create a tuple in Python?

A2. To create a tuple in Python, we can enclose a sequence of elements in parentheses () or use the tuple() function.

Q3. What is the difference between a tuple and a list in Python?

A3. The main difference between a tuple and a list in Python is that tuples are immutable while lists are mutable. This means that we can change the elements of a list, but we cannot change the elements of a tuple once it is created. Additionally, lists are represented using square brackets [], while tuples are represented using parentheses ().

Q4. Can a tuple be changed in Python?

A4. No, a tuple cannot be changed in Python. Once a tuple is created, you cannot add, remove or modify its elements.

Q5. How do you access elements in a tuple?

A5. we can access elements in a tuple using indexing, just like we would with a list. Indexing in Python starts from 0.

Q6. How do you unpack a tuple in Python?

A6. To unpack a tuple in Python, we can assign its elements to multiple variables at once. The number of variables must match the number of elements in the tuple.
