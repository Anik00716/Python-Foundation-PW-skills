Q1. What is a set in Python?

A1. In Python, a set is an unordered collection of unique elements. It is represented using curly braces { } and each element is separated by a comma.

Q2.How do you create a set in Python?

A2. To create a set in Python, we can use the set() function or enclose a list of elements in curly braces {}. For example:

my_set = set()

my_set = {1, 2, 3}

Q3. How do you add elements to a set in Python?

A3. To add elements to a set in Python, we can use the add() method or the update() method. The add() method adds a single element to the set, while the update() method adds multiple elements at once. 

Q4. How do you remove elements from a set in Python?

A4. To remove elements from a set in Python, we can use the remove() method or the discard() method. The remove() method removes a specified element from the set, while the discard() method removes a specified element if it is present in the set, otherwise it does nothing.

Q5. How do you get the length of a set in Python?

A5. To get the length of a set in Python, you can use the len() function.
